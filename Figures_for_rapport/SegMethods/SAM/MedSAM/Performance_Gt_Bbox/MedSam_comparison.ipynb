```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Mean dice for each subject in the test split:

#MedSAM_1C:
#Mean dice for all subjects: 0.544
MedSAM_1C = [0.3063675853137196, 0.8868778280542986, 0.4096048841016995, 0.7006972889956924, 0.5293804251780239, 0.5100019869645489, 0.3908980910844755, 0.6181347112797858]


#MedSAM_3C_DWIADCb0:
#Mean dice for all subjects: 0.607
MedSAM_3C_DWIADCb0 = [0.3616371172135101, 0.8756218905472637, 0.5587840094503158, 0.7271840243653163, 0.6288850038850039, 0.5556364985389969, 0.560964080807526, 0.5890536756660277]


#MedSAM_3C_DWIADCDifDif:
#Mean dice for all subjects: 0.569
MedSAM_3C_DWIADCDifDif = [0.41509118431853625, 0.569620253164557, 0.5128473342945371, 0.6866907078200469, 0.7715258963484115, 0.5516839200273428, 0.4007340375814085, 0.6415591314792591]
```

```python
# X-axis positions, adjust spacing further
ind = np.arange(len(MedSAM_1C)) * 2  # multiply by 2 to increase space between groups
width = 0.5  # the width of the bars

fig, ax = plt.subplots(figsize=(14, 8))  # Increase figure width if necessary
rects1 = ax.bar(ind - width, MedSAM_1C, width, label='MedSAM 1C')
rects2 = ax.bar(ind, MedSAM_3C_DWIADCb0, width, label='MedSAM 3C_DWIADCb0')
rects3 = ax.bar(ind + width, MedSAM_3C_DWIADCDifDif, width, label='MedSAM 3C_DWIADCDifDif')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Index')
ax.set_ylabel('Mean dice for each subject')
ax.set_title('MedSAM comparison')
ax.set_xticks(ind)
ax.set_xticklabels([f'Subject {i+1}' for i in range(len(MedSAM_1C))])
ax.legend()

# Function to add labels on the bars
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(round(height, 2)),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Call the function to add labels
autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.show()
```
